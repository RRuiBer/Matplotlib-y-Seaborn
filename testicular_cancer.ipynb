```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
testicular_cancer = pd.read_csv('/mnt/d/Users/Ricardo/Cursos/6_matplotlib_seaborn/proyecto/Testicular Cancer.csv')
stageI = testicular_cancer.iloc[11].copy()
stageIS  = testicular_cancer.iloc[0].copy()
testicular_cancer.iloc[0] = stageI
testicular_cancer.iloc[11] = stageIS
stageIA = testicular_cancer.iloc[3].copy()
stageIS  = testicular_cancer.iloc[1].copy()
testicular_cancer.iloc[1] = stageIA
testicular_cancer.iloc[3] = stageIS
stageIB = testicular_cancer.iloc[10].copy()
stageIS  = testicular_cancer.iloc[2].copy()
testicular_cancer.iloc[2] = stageIB
testicular_cancer.iloc[10] = stageIS
stageII = testicular_cancer.iloc[50].copy()
stageIS  = testicular_cancer.iloc[4].copy()
testicular_cancer.iloc[4] = stageII
testicular_cancer.iloc[50] = stageIS
stageIIC = testicular_cancer.iloc[148].copy()
stageIS  = testicular_cancer.iloc[10].copy()
testicular_cancer.iloc[10] = stageIIC
testicular_cancer.iloc[148] = stageIS
stageIII = testicular_cancer.iloc[60].copy()
stageIS  = testicular_cancer.iloc[11].copy()
testicular_cancer.iloc[11] = stageIII
testicular_cancer.iloc[60] = stageIS
stageIIIA = testicular_cancer.iloc[46].copy()
stageIS  = testicular_cancer.iloc[12].copy()
testicular_cancer.iloc[12] = stageIIIA
testicular_cancer.iloc[60] = stageIS
testicular_cancer.head(5)
```
